**TASK 01**

**ERROR CODE**

```python
# Bug: Mutable default argument
def add_item(item, items=[]):
    items.append(item)
    return items
print(add_item(1))
print(add_item(2))
```

**FIXED CODE**

```python
# Bug: Mutable default argument
def add_item(item, items=None):
    if items is None:
        items = []
    items.append(item)
    return items
print(add_item(1))
print(add_item(2))
```

**TASK 02**

**ERROR CODE**

```python
# Bug: Floating point precision issue
def check_sum():
    return (0.1 + 0.2) == 0.3
print(check_sum())
```

**FIXED CODE**

```python
def check_sum():
    return abs((0.1 + 0.2) - 0.3) < 1e-9

print(check_sum())
```

**TASK 03**

**ERROR CODE**

```python
def countdown(n):
    print(n)
    return countdown(n-1)

countdown(5)
```

**FIXED CODE**

```python
def countdown(n):
    if n <= 0:   # Base case
        return
    print(n)
    countdown(n-1)
countdown(5)
```

```python
def loop_example():
    i = 0
    while i < 5:
        print(i)
loop_example()
```

**FIXED CODE**

```python
def loop_example():
    i = 0
    while i < 5:
        print(i)
        i += 1   # Increment added
loop_example()
```
