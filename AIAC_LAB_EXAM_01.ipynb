**QUESTION 01**

```python
def count_words_starting_with_a(word_list):
    count = 0
    for word in word_list:
        if word.lower().startswith('a'):
            count += 1
    return count
word_list = ["Apple", "Ball", "Ant", "Cat", "Arrow", "Dog"]
result = count_words_starting_with_a(word_list)
print("Number of words starting with 'A':", result)
```

```python
# Test the function with a sample list
sample_words = ["Apple", "Banana", "Apricot", "Cherry", "Avocado", "grape"]
result = count_words_starting_with_a(sample_words)
print(f"The number of words starting with 'A' is: {result}")
```

```python
def calculate_tax(amount):
    if not isinstance(amount, (int, float)):
        raise TypeError("Amount must be a number.")
    if amount < 0:
        raise ValueError("Amount cannot be negative.")
    return round(amount * 0.05, 2)


try:
    user_input = float(input("Enter the amount: "))
    tax = calculate_tax(user_input)
    print("Calculated Tax:", tax)
except Exception as e:
    print("Error:", e)
```
